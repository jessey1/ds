```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv('train.csv')
```

```python
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic_df)
plt.title('Distribution of Age with respect to Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
sns.barplot(x='Sex',y='Age',data=titanic_df)
plt.show()
```

```python
sns.histplot(data=titanic_df,x='Fare')
```

```python
data = np.random.gamma(2, 1, 1000)  # Example of a right-skewed distribution (Gamma distribution)
```

```python
plt.hist(data, bins='auto')
plt.title('Original Data Histogram')
plt.show()
```

```python
log_data = np.log(data)
```

```python
plt.hist(log_data, bins=200)
plt.title('Normalized Data Histogram (Log Transformation)')
plt.show()
```

```python
sns.catplot(x="Age", y="Survived", data=titanic_df)
plt.show()
```
